```python
import os
import re

# ==========================
# MAIN DATASET PATH
# ==========================
base_dir = "../Data_prep"

# Supported image extensions
img_exts = (".jpg", ".jpeg", ".png")

# ==========================
# RENAME FUNCTION (SAFE)
# ==========================
def rename_images(folder_path, prefix):
    files = sorted([
        f for f in os.listdir(folder_path)
        if f.lower().endswith(img_exts)
    ])

    counter = 1
    renamed = 0
    skipped = 0

    for filename in files:
        # If already renamed â†’ skip
        if re.match(rf"{prefix}_\d+\.(jpg|jpeg|png)", filename, re.IGNORECASE):
            skipped += 1
            continue

        old_path = os.path.join(folder_path, filename)
        ext = os.path.splitext(filename)[1]
        new_name = f"{prefix}_{counter}{ext}"
        new_path = os.path.join(folder_path, new_name)

        # Avoid overwrite
        while os.path.exists(new_path):
            counter += 1
            new_name = f"{prefix}_{counter}{ext}"
            new_path = os.path.join(folder_path, new_name)

        os.rename(old_path, new_path)
        counter += 1
        renamed += 1

    print(f"{prefix}: Renamed={renamed}, Skipped={skipped}")

# ==========================
# PROCESS EACH CLASS FOLDER
# ==========================
for class_name in ["BCC", "BKL", "NV", "MEL"]:
    folder = os.path.join(base_dir, class_name)
    if os.path.exists(folder):
        rename_images(folder, class_name)
    else:
        print(f"Folder not found: {class_name}")
```
